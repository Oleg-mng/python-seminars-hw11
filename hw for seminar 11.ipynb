```python
import sympy
x=sympy.symbols('x')
print(sympy.solve(5*x**2+10*x-30))
```

```python
import sympy
x=sympy.symbols('x')
f=5*x**2+10*x-30
koren_x=sympy.solve(f.diff(x))
koren_y=5*koren_x[0]**2+10*koren_x[0]-30
print(f'Корни квадратного уравнения ({koren_x[0]},{koren_y})')
```

```python
print(f'Функция убывает на промежутке (-∞, {koren_x[0]})')
```

```python
print(f'Функция возрастает на промежутке ({koren_x[0]}, +∞)')
```

```python
print(f'Вершина функции в точке ({koren_x[0]},{koren_y})')
```

```python
import sympy
x=sympy.symbols('x')
f=-12*x**4*sympy.sin(sympy.cos(x)) - 18*x**3+5*x**2 + 10*x - 30
koren_x=sympy.solve(f.diff(x))
koren_y=-12*koren_x**4*sympy.sin(sympy.cos(koren_x)) - 18*koren_x**3+5*koren_x**2 + 10*koren_x - 30
print(f'Корни квадратного уравнения ({koren_x[0]},{koren_y})')
```
